```python
import datetime as dt

# Convert 'event_time' to datetime if not already done
df['event_time'] = pd.to_datetime(df['event_time'])

# Define the churn threshold (e.g., 30 days)
latest_event_time = df['event_time'].max()
churn_threshold = latest_event_time - dt.timedelta(days=30)

# Identify the last event time for each user
user_last_event = df.groupby('user_id')['event_time'].max()

# Mark churned users
churned_users = user_last_event[user_last_event < churn_threshold]
active_users = user_last_event[user_last_event >= churn_threshold]

print(f"Total Users: {len(user_last_event)}")
print(f"Churned Users: {len(churned_users)}")
print(f"Active Users: {len(active_users)}")
```

```python
# Calculate percentages
user_status = pd.Series([len(active_users), len(churned_users)], index=['Active', 'Churned'])

# Plot pie chart
user_status.plot(kind='pie', autopct='%1.1f%%', startangle=140, title='User Status')
plt.ylabel('')
plt.show()
```

```python
# Filter data for churned users
churned_user_data = df[df['user_id'].isin(churned_users.index)]

# Top categories for churned users
churned_categories = churned_user_data['category_code'].value_counts().head(10)
print("Top Categories for Churned Users:\n", churned_categories)

# Top brands for churned users
churned_brands = churned_user_data['brand'].value_counts().head(10)
print("Top Brands for Churned Users:\n", churned_brands)

# Plot churned categories
churned_categories.plot(kind='bar', title='Top Categories for Churned Users', color='red')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Plot churned brands
churned_brands.plot(kind='bar', title='Top Brands for Churned Users', color='orange')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
# Purchases by churned users
churned_purchases = churned_user_data[churned_user_data['event_type'] == 'purchase']
avg_purchase_value_churned = churned_purchases['price'].mean()
print(f"Average Purchase Value (Churned): {avg_purchase_value_churned}")

# Purchases by active users
active_user_data = df[df['user_id'].isin(active_users.index)]
active_purchases = active_user_data[active_user_data['event_type'] == 'purchase']
avg_purchase_value_active = active_purchases['price'].mean()
print(f"Average Purchase Value (Active): {avg_purchase_value_active}")
```

```python
from sklearn.model_selection import train_test_split

# Create a feature set for each user
user_features = df.groupby('user_id').agg(
    total_events=('event_type', 'count'),
    total_purchases=('event_type', lambda x: (x == 'purchase').sum()),
    total_views=('event_type', lambda x: (x == 'view').sum()),
    total_cart=('event_type', lambda x: (x == 'cart').sum()),
    average_price=('price', 'mean'),
    favorite_category=('category_code', lambda x: x.mode()[0] if not x.mode().empty else 'unknown')
).reset_index()

# Merge churn labels
user_features['churned'] = user_features['user_id'].apply(lambda x: 1 if x in churned_users.index else 0)

# Convert categorical features to numerical
user_features = pd.get_dummies(user_features, columns=['favorite_category'], drop_first=True)

# Separate features and target variable
X = user_features.drop(columns=['user_id', 'churned'])
y = user_features['churned']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training samples: {len(X_train)}, Testing samples: {len(X_test)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train logistic regression model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Make predictions
y_pred = lr_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train random forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# Feature importance from Random Forest
import matplotlib.pyplot as plt
import seaborn as sns

feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)

# Plot feature importance
sns.barplot(x=feature_importances.values, y=feature_importances.index)
plt.title('Feature Importance')
plt.show()
```
